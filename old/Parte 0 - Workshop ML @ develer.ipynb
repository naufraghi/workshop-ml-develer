# Workshop Develer - Introduzione al Machine Learning

Il machine learning consiste nel creare programmi con **parametri modificabili** (tipicamente una array di float) che vengono scelti automaticamente in modo da migliorare la capacità del programma di **adattarsi a nuovi dati**.

L'apprendimento automatico può essere visto come una branca dell'**Intelligenza Artificiale** dato gli algoritmi di apprendimento possono essere visti come componenti di un computer con un comportamento intelligente, capace in qualche modo di **generalizzare**, cosa diversa dal memorizzare e ritrovare dati come farebbe un database.

Se avete letto [Machine Learning is Fun!](https://medium.com/@ageitgey/machine-learning-is-fun-80ea3ec3c471) immagino avrete voglia di fare qualche esperimento in prima persona.

Avete visto che ci sono (almeno) due tipi di apprendimento automatico:

- l'apprendimento supervisionato e
- l'apprendimento non supervisionato

in questo workshop ci occuperemo solo di apprendimento supervisionato, in particolare di regressione e classificazione.
